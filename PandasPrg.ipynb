```python
import pandas as pd
import numpy as np

# 1) Create Series
a = [1, 2, 3, 4]
b = pd.Series(a)
print("1) Series created from list a:", b)

# 2) Create index on Series
a = [1, 2, 3]
b = pd.Series(a, index=["x", "y", "z"])
print("2) Series with custom index:", b)

# 3) head
print("3) Head of the Series (first 2 elements):", b.head(2))

# 4) tail
print("4) Tail of the Series (last 2 elements):", b.tail(2))

# 5) Get index
print("5) Index of the Series:", b.index)

# 6) Get values
print("6) Values of the Series:", b.values)

# 7) Sort values
print("7) Sorted values of the Series:", b.sort_values())

# 8) Data type of Series
print("8) Data type of the Series:", b.dtype)

# 9) Change the data type
print("9) Series converted to float type:", b.astype(float))

# 10) Sum
print("10) Sum of the Series:", b.sum())

# 11) Min
print("11) Minimum value in the Series:", b.min())

# 12) Max
print("12) Maximum value in the Series:", b.max())

# 13) Mean
print("13) Mean of the Series:", b.mean())

# 14) Median
print("14) Median of the Series:", b.median())

# 15) Standard deviation
print("15) Standard deviation of the Series:", b.std())

# 16) Drop NaN values
ser = pd.Series([1.1, 2.2, np.nan])
dropSer = ser.dropna()
print("16) Series after dropping NaN values:", dropSer)

# 17) Fill NaN values
ser = pd.Series([1.1, 2.2, np.nan])
ser.fillna(0, inplace=True)
print("17) Series after filling NaN values with 0:", ser)

# 18) Series to dictionary
s = pd.Series([1, 2, 3, 4])
print("18) Series converted to dictionary:", s.to_dict())

# 19) Create DataFrame
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}
d = pd.DataFrame(data)
print("19) DataFrame created from data:", d)

# 20) Add index to DataFrame
d_indexed = pd.DataFrame(data, index=["Round1", "Round2", "Round3"])
print("20) DataFrame with custom index:", d_indexed)

# 21) Display columns
print("21) Columns in the DataFrame:", d.columns)

# 22) Rename columns
c = d.rename(columns={'duration': 'time'})
print("22) DataFrame after renaming column 'duration' to 'time':", c)

# 23) Drop column
c = d.drop(columns=['duration'])
print("23) DataFrame after dropping 'duration' column:", c)

# 24) Add to DataFrame
data1 = {
    "col1": [10, 20],
    "col2": [30, 40]
}
df = pd.DataFrame(data1)
c = df.add(1)
print("24) DataFrame after adding 1 to each element:", c)

# 25) Subtract from DataFrame
c = df.sub(1)
print("25) DataFrame after subtracting 1 from each element:", c)

# 26) Multiply DataFrame
c = df.mul(2)
print("26) DataFrame after multiplying each element by 2:", c)

# 27) Modulo operation on DataFrame
c = df.mod(10)
print("27) DataFrame after performing modulo 10 on each element:", c)

# 28) Power operation on DataFrame
c = df.pow(3)
print("28) DataFrame after raising each element to the power of 3:", c)

# 29) Equal to operation
c = df.eq(10)
print("29) DataFrame showing equality to 10:", c)

# 30) Greater than operation
c = df.gt(20)
print("30) DataFrame showing greater than 20:", c)

# 31) Less than or equal to operation
c = df.le(40)
print("31) DataFrame showing less than or equal to 40:", c)

# 32) Transform operation (using lambda function)
c = df.transform(lambda x: x + 3)
print("32) DataFrame after transforming with lambda (adding 3):", c)

# 33) Correlation between features
print("33) Correlation between features in DataFrame:\n", df.corr())

# 34) Covariance between features
print("34) Covariance between features in DataFrame:\n", df.cov())

# 35) Count of non-null observations
print("35) Count of non-null observations in DataFrame:\n", df.count())
```
