```python
import pandas as pd
import numpy as np

# 1) Create Series
a = [1, 2, 3, 4]
b = pd.Series(a)
print(b)

# 2) Create index on Series
a = [1, 2, 3]
b = pd.Series(a, index=["x", "y", "z"])
print(b)

# 3) head
print(b.head(2))

# 4) tail
print(b.tail(2))

# 5) Get index
print(b.index)

# 6) Get values
print(b.values)

# 7) Sort values
print(b.sort_values())

# 8) Data type of Series
print(b.dtype)

# 9) Change the data type
print(b.astype(float))

# 10) Sum
print(b.sum())

# 11) Min
print(b.min())

# 12) Max
print(b.max())

# 13) Mean
print(b.mean())

# 14) Median
print(b.median())

# 15) Standard deviation
print(b.std())

# 16) Drop NaN values
ser = pd.Series([1.1, 2.2, np.nan])
dropSer = ser.dropna()
print(dropSer)

# 17) Fill NaN values
ser = pd.Series([1.1, 2.2, np.nan])
ser.fillna(0, inplace=True)
print(ser)

# 18) Series to dictionary
s = pd.Series([1, 2, 3, 4])
print(s.to_dict())

# 19) Create DataFrame
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}
d = pd.DataFrame(data)
print(d)

# 20) Add index to DataFrame
d_indexed = pd.DataFrame(data, index=["Round1", "Round2", "Round3"])
print(d_indexed)

# 21) Display columns
print(d.columns)

# 22) Rename columns
c = d.rename(columns={'duration': 'time'})
print(c)

# 23) Drop column
c = d.drop(columns=['duration'])
print(c)

# 24) Add to DataFrame
data1 = {
    "col1": [10, 20],
    "col2": [30, 40]
}
df = pd.DataFrame(data1)
c = df.add(1)
print(c)

# 25) Subtract from DataFrame
c = df.sub(1)
print(c)

# 26) Multiply DataFrame
c = df.mul(2)
print(c)

# 27) Modulo operation on DataFrame
c = df.mod(10)
print(c)

# 28) Power operation on DataFrame
c = df.pow(3)
print(c)

# 29) Equal to operation
c = df.eq(10)
print(c)

# 30) Greater than operation
c = df.gt(20)
print(c)

# 31) Less than or equal to operation
c = df.le(40)
print(c)

# 32) Transform operation (using lambda function)
c = df.transform(lambda x: x + 3)
print(c)

# 33) Correlation between features
print(df.corr())

# 34) Covariance between features
print(df.cov())

# 35) Count of non-null observations
print(df.count())
```
