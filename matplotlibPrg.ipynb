```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

#1) plot
xpoints = [0, 6]
ypoints = [0, 250]
plt.plot(xpoints, ypoints)

#2) plot without line
plt.plot(xpoints, ypoints, 'o')

#3) marker
plt.plot(ypoints, marker="*")

#4) marker size
plt.plot(ypoints, marker='o', ms=35)

#5) linestyle
plt.plot(ypoints, linestyle='dotted')

#6) labels
plt.plot(xpoints, ypoints)
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

#7) title
plt.title("Sports Watch Data")

#8) grid lines to plot
plt.grid()

#9) subplots
plt.subplot(1, 2, 1)
plt.plot(xpoints, ypoints)
plt.subplot(2, 2, 2)
plt.plot(ypoints, xpoints)

#10) scatter plot
plt.scatter(xpoints, ypoints)

#11) creating bar
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y)

#12) horizontal bar
plt.barh(x, y)

#13) histogram
x = np.random.normal(170, 10, 250)
plt.hist(x)

#14) pie charts
y = np.array([35, 25, 25, 15])
plt.pie(y)

#15) labels to pie chart
mylabels = ["Apple", "Banana", "Cherries", "Dates"]
plt.pie(y, labels=mylabels)

#16) start angle
plt.pie(y, startangle=90)

#17) box plot
data = np.random.rand(10, 4)
plt.boxplot(data)

#18) violin plot
data = np.random.rand(10, 4)
plt.violinplot(data)

#19) heatmap (corrected interpolation spelling)
data = np.random.rand(10, 10)
plt.imshow(data, cmap="hot", interpolation="nearest")
plt.colorbar()

#20) 3D plots
fig = plt.figure()
ax = plt.axes(projection='3d')

#21) line on 3D plot (added x, y, z)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = np.sin(np.sqrt(x**2 + y**2))
ax.plot3D(x, y, z)

#22) save the image
plt.savefig('File Path')

#23) semilogx
plt.semilogx(x, y, marker=".")

#24) semilogy
plt.semilogy(x, y, marker=".")

#25) stem plot
#plt.stem([0.1, 0.2, 0.3, 0.4], [1, 2, 3, 4])
#plt.show()

#26) step plot (corrected from stem to step)
plt.step([1, 2, 3, 4], [10, 20, 15, 25])
plt.show()

#27) xlim
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlim(0, 3)

#28) ylim (corrected xlim to ylim)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylim(0, 18)

#29) axhline
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axhline(y=10, color='r')

#30) axvline
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axvline(x=2.5, color='r')

#31) margins
plt.margins(0.1)

#32) add legend
plt.legend(['line1'])

#33) show figure
plt.figure()

#34) fill between two lines
plt.fill_between([1, 2, 3], [2, 3, 4], [1, 2, 3])

#35) add text to position
plt.text(2, 3, 'Point')

#36) change linewidth
plt.plot([1, 2, 3], [4, 5, 6], linewidth=2)

#37) linestyle (missing example, left blank for user input)

#38) errorbar
plt.errorbar([1, 2, 3], [4, 5, 6], yerr=0.2)

#39) colorbar
data = np.random.rand(10, 10)
img = plt.imshow(data, cmap="viridis")
plt.colorbar(img)
plt.show()

#40) Polar plot
plt.polar([0, 1, 2], [1, 2, 3])

plt.show()
```
