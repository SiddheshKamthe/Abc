```python
import numpy as np

# 1) Dimension of an array
arr2 = np.array([[1, 2], [3, 4]])
print("1) Dimension of array:", arr2.ndim)

# 2) Type of array
print("2) Type of array:", type(arr2))

# 3) Slicing array
arr5 = np.array([1, 2, 3, 4, 5, 6, 7])
print("3) Slicing array:", arr5[1:5:2])

# 4) Shape of matrix
print("4) Shape of matrix:", arr2.shape)

# 5) Reshape the matrix
newarr = arr2.reshape(1, 4)
print("5) Reshaped matrix:", newarr)

# 6) Joining arrays
a1 = np.array([1, 2])
a2 = np.array([3, 4])
newarr = np.concatenate((a1, a2))
print("6) Concatenated array:", newarr)

# 7) Sorting an array
arr3 = np.array([6, 2, 0, 4])
sorted_arr = np.sort(arr3)
print("7) Sorted array:", sorted_arr)

# 8) Random integer array
arr4 = np.random.randint(0, 100, size=(5))
print("8) Random integer array:", arr4)

# 9) Addition
arr26 = [1, 3, 5]
arr27 = [2, 4, 6]
print("9) Addition:", np.add(arr26[0], arr27[0]))

# 10) Subtraction
print("10) Subtraction:", np.subtract(arr26[0], arr27[0]))

# 11) Modulo
print("11) Modulo:", np.mod(arr26[0], arr27[0]))

# 12) Remainder
print("12) Remainder:", np.remainder(arr26[0], arr27[0]))

# 13) Power
print("13) Power:", np.power(arr26[0], arr27[0]))

# 14) Exponent
print("14) Exponent:", np.exp(arr27[0]))

# 15) Square root
r33 = [1, 16, 9]
print("15) Square root:", np.sqrt(r33))

# 16) Logarithm
print("16) Logarithm:", np.log(r33))

# 17) Ceiling
arr35 = [1.66, 1.734, 6.44]
print("17) Ceil:", np.ceil(arr35))

# 18) Floor
print("18) Floor:", np.floor(arr35))

# 19) Minimum
print("19) Min:", np.min(arr35))

# 20) Maximum
print("20) Max:", np.max(arr35))

# 21) Sum
print("21) Sum:", np.sum(arr35))

# 22) Sine
arr = np.array([0, np.pi/2, np.pi])
print("22) Sine:", np.sin(arr))

# 23) Cosine
print("23) Cosine:", np.cos(arr))

# 24) Tangent
print("24) Tangent:", np.tan(arr))

# 25) Inverse Sine
print("25) Inverse Sine:", np.arcsin(np.clip(arr / np.pi, -1, 1)))

# 26) Inverse Cosine
print("26) Inverse Cosine:", np.arccos(np.clip(arr / np.pi, -1, 1)))

# 27) Zeros array
arr50 = np.zeros((2, 2))
print("27) Zeros array:\n", arr50)

# 28) Ones array
arr50 = np.ones((2, 2))
print("28) Ones array:\n", arr50)

# 29) Mean
arr69 = [1, 2, 3, 4, 5]
print("29) Mean:", np.mean(arr69))

# 30) Median
print("30) Median:", np.median(arr69))

# 31) Standard deviation
print("31) Standard deviation:", np.std(arr69))

# 32) Variance
print("32) Variance:", np.var(arr69))

# 33) Product of array
print("33) Product:", np.prod(arr69))

# 34) Append to array
arr59 = np.array([1, 2, 3])
b = np.append(arr59, [4, 5, 6])
print("34) Appended array:", b)

# 35) Insert into array
b = np.insert(arr59, 1, [10, 20])
print("35) Inserted array:", b)
```
