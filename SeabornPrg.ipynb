```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Load dataset
penguins = sns.load_dataset("penguins")

# 2) Barplot
sns.barplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 3) Countplot
sns.countplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 4) Histplot
sns.histplot(data=penguins, x='flipper_length_mm', hue="species", multiple="stack")
plt.show()

# 5) Kernel Density Estimation plot (using histplot)
sns.histplot(data=penguins, x='flipper_length_mm', hue="species", kde=True)
plt.show()

# 6) Creating histplot using displot
sns.displot(data=penguins, x='flipper_length_mm', hue="species", kind="hist")
plt.show()

# 7) Scatter plot
tip = sns.load_dataset("tips")
sns.scatterplot(data=tip, x="total_bill", y="tip")
plt.show()

# 8) Relationship plot
sns.relplot(data=tip, x="total_bill", y="tip")
plt.show()

# 9) Pairplot
sns.pairplot(data=penguins, hue="species")
plt.show()

# 10) FacetGrid
g = sns.FacetGrid(penguins)
plt.show()

# 11) Head of flights dataset
flight = sns.load_dataset("flights")
print(flight.head())

# 12) Tail of flights dataset
print(flight.tail())

# 13) Stripplot
sns.stripplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 14) Set theme
sns.set_theme(style="darkgrid")

# 15) Set style
sns.set_style("whitegrid")

# 16) Set context
sns.set_context("paper")

# 17) Set palette
sns.set_palette("husl")

# 18) Color palette
palette = sns.color_palette("bright")
sns.palplot(palette)
plt.show()

# 19) Palplot
sns.palplot(sns.color_palette("Paired"))
plt.show()

# 20) Get dataset names
print(sns.get_dataset_names())

# 21) Set X label
ax = sns.scatterplot(data=tip, x="total_bill", y="tip")
ax.set_xlabel("Total Bill Amount")
plt.show()

# 22) Set Y label
ax.set_ylabel("Tip Amount")
plt.show()

# 23) Set X ticks
ax.set_xticks([10, 20, 30, 40, 50])

# 24) Set Y ticks
ax.set_yticks([1, 2, 3, 4, 5])

# 25) Set X tick labels
ax.set_xticklabels(['Very Low', 'Low', 'Medium', 'High', 'Very High'], rotation=45)

# 26) Set Y tick labels
ax.set_yticklabels(['Very Low', 'Low', 'Medium', 'High', 'Very High'], rotation=45)

# 27) Set X limits
ax.set_xlim(left=0, right=60)

# 28) Set Y limits
ax.set_ylim(bottom=0, top=12)

# 29) Add grid
ax.grid(True)

# 30) Remove grid
ax.grid(False)

# 31) Lineplotpenguins = sns.load_dataset("penguins")

# Create a line plot for flipper_length_mm by species
sns.lineplot(data=penguins, x="species", y="flipper_length_mm", estimator='mean')
plt.show()

# 32) Boxplot
iris = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()

# 33) Violinplot
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.show()

# 34) Light palette
sns.light_palette("green")

# 35) Dark palette
sns.dark_palette("purple")
```
