```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Load dataset
print("Loading penguins dataset...")
penguins = sns.load_dataset("penguins")

# 2) Barplot
print("Creating a barplot for flipper lengths by species...")
sns.barplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 3) Countplot
print("Creating a countplot for flipper lengths by species...")
sns.countplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 4) Histplot
print("Creating a histplot for flipper lengths by species...")
sns.histplot(data=penguins, x='flipper_length_mm', hue="species", multiple="stack")
plt.show()

# 5) Kernel Density Estimation plot (using histplot)
print("Creating a histplot with KDE for flipper lengths by species...")
sns.histplot(data=penguins, x='flipper_length_mm', hue="species", kde=True)
plt.show()

# 6) Creating histplot using displot
print("Creating a displot for flipper lengths by species...")
sns.displot(data=penguins, x='flipper_length_mm', hue="species", kind="hist")
plt.show()

# 7) Scatter plot
print("Loading tips dataset for scatter plot...")
tip = sns.load_dataset("tips")
print("Creating a scatter plot for total bill vs tip...")
sns.scatterplot(data=tip, x="total_bill", y="tip")
plt.show()

# 8) Relationship plot
print("Creating a relationship plot for total bill vs tip...")
sns.relplot(data=tip, x="total_bill", y="tip")
plt.show()

# 9) Pairplot
print("Creating a pairplot for penguins dataset...")
sns.pairplot(data=penguins, hue="species")
plt.show()

# 10) FacetGrid
print("Creating a FacetGrid for penguins dataset...")
g = sns.FacetGrid(penguins)
plt.show()

# 11) Head of flights dataset
print("Loading flights dataset and displaying the head...")
flight = sns.load_dataset("flights")
print(flight.head())

# 12) Tail of flights dataset
print("Displaying the tail of the flights dataset...")
print(flight.tail())

# 13) Stripplot
print("Creating a stripplot for flipper lengths by species...")
sns.stripplot(data=penguins, x='flipper_length_mm', hue="species")
plt.show()

# 14) Set theme
print("Setting the Seaborn theme to darkgrid...")
sns.set_theme(style="darkgrid")

# 15) Set style
print("Setting the Seaborn style to whitegrid...")
sns.set_style("whitegrid")

# 16) Set context
print("Setting the Seaborn context to paper...")
sns.set_context("paper")

# 17) Set palette
print("Setting the Seaborn color palette to husl...")
sns.set_palette("husl")

# 18) Color palette
print("Creating and displaying a bright color palette...")
palette = sns.color_palette("bright")
sns.palplot(palette)
plt.show()

# 19) Palplot
print("Displaying the Paired color palette...")
sns.palplot(sns.color_palette("Paired"))
plt.show()

# 20) Get dataset names
print("Getting dataset names available in Seaborn...")
print(sns.get_dataset_names())

# 21) Set X label
print("Creating a scatter plot and setting X label...")
ax = sns.scatterplot(data=tip, x="total_bill", y="tip")
ax.set_xlabel("Total Bill Amount")
plt.show()

# 22) Set Y label
print("Setting Y label for scatter plot...")
ax.set_ylabel("Tip Amount")
plt.show()

# 23) Set X ticks
print("Setting X ticks for scatter plot...")
ax.set_xticks([10, 20, 30, 40, 50])

# 24) Set Y ticks
print("Setting Y ticks for scatter plot...")
ax.set_yticks([1, 2, 3, 4, 5])

# 25) Set X tick labels
print("Setting X tick labels for scatter plot...")
ax.set_xticklabels(['Very Low', 'Low', 'Medium', 'High', 'Very High'], rotation=45)

# 26) Set Y tick labels
print("Setting Y tick labels for scatter plot...")
ax.set_yticklabels(['Very Low', 'Low', 'Medium', 'High', 'Very High'], rotation=45)

# 27) Set X limits
print("Setting X limits for scatter plot...")
ax.set_xlim(left=0, right=60)

# 28) Set Y limits
print("Setting Y limits for scatter plot...")
ax.set_ylim(bottom=0, top=12)

# 29) Add grid
print("Adding grid to scatter plot...")
ax.grid(True)

# 30) Remove grid
print("Removing grid from scatter plot...")
ax.grid(False)

# 31) Lineplot
print("Creating a line plot for flipper_length_mm by species...")
sns.lineplot(data=penguins, x="species", y="flipper_length_mm", estimator='mean')
plt.show()

# 32) Boxplot
print("Creating a boxplot for sepal lengths in iris dataset...")
iris = sns.load_dataset("iris")
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()

# 33) Violinplot
print("Creating a violin plot for sepal lengths in iris dataset...")
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.show()

# 34) Light palette
print("Creating a light color palette...")
sns.light_palette("green")

# 35) Dark palette
print("Creating a dark color palette...")
sns.dark_palette("purple")
```
